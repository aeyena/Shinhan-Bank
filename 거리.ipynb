```python
import pandas as pd

# 은행 군집 데이터 파일 가져오기
df = pd.read_csv("C:/Users/yena/Documents/대학4-1/전공종합설계및논문/data/bank_clustering.csv", encoding = 'euc-kr')
df.head()
```

```python
# 은행 군집 0,3,4,5,6번 추출

df = df[df['cluster'] != 1]
bank_df = df[df['cluster'] != 2]
bank_df.reset_index(drop=True, inplace=True)
bank_df.head()
```

```python
# 지하철 군집 데이터 파일 가져오기 
subway_df = pd.read_csv("C:/Users/yena/Documents/대학4-1/전공종합설계및논문/data/subway_clustering_center.csv")
subway_df.head()
```

```python
# 버스정류장 군집 데이터 파일 가져오기 
bus_df = pd.read_csv("C:/Users/yena/Documents/대학4-1/전공종합설계및논문/data/busstop_clustering_center.csv")
bus_df.head()
```

```python
# 자동화기기 군집 데이터 파일 가져오기 
shinhan_df = pd.read_csv("C:/Users/yena/Documents/대학4-1/전공종합설계및논문/data/자동화기기_군집중심.csv")
shinhan_df.head()
```

```python
# 경로당 군집 데이터 파일 가져오기 
senior_df = pd.read_csv("C:/Users/yena/Documents/대학4-1/전공종합설계및논문/data/senior_clustering_center.csv")
senior_df.head()
```

```python
# 시장 군집 데이터 파일 가져오기 
traditional_market_df = pd.read_csv("C:/Users/yena/Documents/대학4-1/전공종합설계및논문/data/전통시장현황_clustering_center.csv")
traditional_market_df.head()
```

```python
from haversine import haversine

# 가장 가까운 지하철 군집과의 거리 구하기 
subway_dist = []

for i in range(len(bank_df)):
    m = 1000000000000000000
    for j in range(len(subway_df)):
        new_m = haversine((bank_df.loc[i]['Y'],bank_df.loc[i]['X']), (subway_df.loc[j]['Y'], subway_df.loc[j]['X']), unit = 'm')
        if  new_m <= m:
            m = new_m
        else:
            continue
    subway_dist.append(m)
    
print(subway_dist)
```

```python
# 가장 가까운 버스정류장 군집과의 거리 구하기 
bus_dist = []

for i in range(len(bank_df)):
    m = 1000000000000000000
    for j in range(len(bus_df)):
        new_m = haversine((bank_df.loc[i]['Y'],bank_df.loc[i]['X']), (bus_df.loc[j]['Y'], bus_df.loc[j]['X']), unit = 'm')
        if  new_m <= m:
            m = new_m
        else:
            continue
    bus_dist.append(m)
    
print(bus_dist)
```

```python
# 가장 가까운 자동화기기 군집과의 거리 구하기 
shinhan_dist = []

for i in range(len(bank_df)):
    m = 1000000000000000000
    for j in range(len(shinhan_df)):
        new_m = haversine((bank_df.loc[i]['Y'],bank_df.loc[i]['X']), (shinhan_df.loc[j]['Y'], shinhan_df.loc[j]['X']), unit = 'm')
        if  new_m <= m:
            m = new_m
        else:
            continue
    shinhan_dist.append(m)
    
print(shinhan_dist)
```

```python
# 가장 가까운 경로당 군집과의 거리 구하기 
senior_dist = []

for i in range(len(bank_df)):
    m = 1000000000000000000
    for j in range(len(senior_df)):
        new_m = haversine((bank_df.loc[i]['Y'],bank_df.loc[i]['X']), (senior_df.loc[j]['Y'], senior_df.loc[j]['X']), unit = 'm')
        if  new_m <= m:
            m = new_m
        else:
            continue
    senior_dist.append(m)
    
print(senior_dist)
```

```python
# 가장 가까운 전통시장 군집과의 거리 구하기 
traditional_market_dist = []

for i in range(len(bank_df)):
    m = 1000000000000000000
    for j in range(len(traditional_market_df)):
        new_m = haversine((bank_df.loc[i]['Y'],bank_df.loc[i]['X']), (traditional_market_df.loc[j]['Y'], traditional_market_df.loc[j]['X']), unit = 'm')
        if  new_m <= m:
            m = new_m
        else:
            continue
    traditional_market_dist.append(m)
    
print(traditional_market_dist)
```

```python
# 기존 데이터 프레임에 각 군집과의 최소거리 열 추가
bank_df['subway_dist'] = subway_dist
bank_df['bus_dist'] = bus_dist
bank_df['shinhan_dist'] = shinhan_dist
bank_df['senior_dist'] = senior_dist
bank_df['traditional_market_dist'] = traditional_market_dist

bank_df
```
