```python
import pandas as pd

# 파일 불러오기
df = pd.read_csv("C:/Users/yena/Documents/대학4-1/전공종합설계및논문/data/국가철도공단_부산_지하철_주소데이터.csv")
df.head()
```

```python
# 필요한 데이터만 추출
df = df[['도로명주소','Y','X']]
subway = df[['Y','X']]
subway.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# 스케일링 진행 
scalar = MinMaxScaler()
subway_scale = scalar.fit_transform(subway)
```

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans

# 최적의 k값 찾기 -> Elbow Method 이용
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,100))
visualizer.fit(subway_scale)
```

```python
k = 11

# 그룹 수, random_state 설정
model = KMeans(n_clusters = k, random_state = 10)

# 스케일링된 데이터에 학습 
model.fit(subway_scale)

# 클러스터링 결과 각 데이터가 몇 번째 그룹에 속하는지 저장
df['cluster'] = model.fit_predict(subway_scale)
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score

# 실루엣 계수 확인
df_silhouette = df[['Y','X','cluster']]
score_samples = silhouette_samples(df_silhouette,df_silhouette['cluster'])

df_silhouette['silhouette_coeff'] = score_samples
average_score=silhouette_score(df_silhouette,df_silhouette['cluster'])

print('군집 개수 %d개의 평균 실루엣: '%k,average_score)
```

```python
from yellowbrick.cluster import SilhouetteVisualizer

#visualizer 생성
visualizer_4 = SilhouetteVisualizer(model, colors='yellowbrick')
#생성된 visualizer에 데이터 입력 
visualizer_4.fit(subway_scale)      
visualizer_4.show()
```

```python
import matplotlib.pyplot as plt

# k-means 군집화 결과 시각화 
plt.figure(figsize = (8,8))

colorcode = ['b','g','r','c','m','k','tab:orange','tab:purple','tab:brown','tab:pink','tab:gray','tab:olive','tab:cyan','darkgreen','indigo','navy','aqua']

for i in range(k):
    plt.scatter(df.loc[df['cluster'] == i, 'X'], df.loc[df['cluster'] == i, 'Y'], label = 'cluster' + str(i), c=colorcode[i])

plt.legend()
plt.title('K = %d results'%k, size = 15)
plt.xlabel('longitude', size = 12)
plt.ylabel('latitude', size = 12)
plt.show()
```

```python
# 각 cluster의 중심 좌표 구하기
labels = model.labels_
centers = model.cluster_centers_

subway_centers = pd.DataFrame(centers, columns=['latitude','longitude'])
subway_centers
```

```python
# 각 cluser의 중심좌표(스케일링 전)
inverse_subway_centers = scalar.inverse_transform(subway_centers)
inverse_subway_centers
```
